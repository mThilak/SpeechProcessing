```python
import numpy as np
import matplotlib
from matplotlib import pyplot as plt

x_train = np.arange(0, 2*np.pi, 2*np.pi/10)
noise10 = np.random.normal(0, 0.2, 10) # noise of 0 mean, 0.2 standard deviation and 10 samples
y_train = np.sin(x_train) + noise10
```

```python
x_test = np.arange(0, 2*np.pi, 2*np.pi/100)
noise100 = np.random.normal(0, 0.2, 100)
y_test = np.sin(x_test) + noise100
```

```python
#Use the following functions:
degree_0= np.polyfit(x_train,y_train,0)
p0=np.poly1d(degree_0)
degree_2= np.polyfit(x_train,y_train,2)
p2=np.poly1d(degree_2)
degree_4= np.polyfit(x_train,y_train,4)
p4=np.poly1d(degree_4)
degree_6= np.polyfit(x_train,y_train,6)
p6=np.poly1d(degree_6)
degree_9= np.polyfit(x_train,y_train,9)
p9=np.poly1d(degree_9)
```

```python
plt.plot(x_train,p4(x_train),"red")
plt.plot(x_train,y_train,"blue")
plt.show()
```

```python
#np.sqrt, np.mean for residual
#PLot degree versus training and test error
from sklearn.metrics import mean_squared_error
from math import sqrt
train0 = sqrt(mean_squared_error(p0(x_train), y_train))
train2 = sqrt(mean_squared_error(p2(x_train), y_train))
train4 = sqrt(mean_squared_error(p4(x_train), y_train))
train6 = sqrt(mean_squared_error(p6(x_train), y_train))
train9 = sqrt(mean_squared_error(p9(x_train), y_train))
degree=np.array([0,2,4,6,9])
trainarray=np.array([train0,train2,train4,train6,train9])
test0 = sqrt(mean_squared_error(p0(x_test), y_test))
test2 = sqrt(mean_squared_error(p2(x_test), y_test))
test4 = sqrt(mean_squared_error(p4(x_test), y_test))
test6 = sqrt(mean_squared_error(p6(x_test), y_test))
test9 = sqrt(mean_squared_error(p9(x_test), y_test))
testarray=np.array([test0,test2,test4,test6,test9])

plt.plot(degree,trainarray,'red')
plt.plot(degree,testarray,'blue')
plt.show()
```

```python
from sklearn import linear_model
from sklearn.linear_model import Ridge
from sklearn.preprocessing import PolynomialFeatures 
from sklearn.pipeline import make_pipeline
from sklearn.metrics import mean_squared_error
```

```python

x_train1 = np.linspace(0, 1, 10)
noise = np.random.normal(0, 0.2, 10)
y_train1 = np.sin(2*np.pi*x_train1) + noise
x_train1 = x_train1[:, np.newaxis] #dont run more than once it goes on adding axis

x_test1 = np.linspace(0, 1, 10)
noise = np.random.normal(0, 0.2, 10)
y_test1 = np.sin(2*np.pi*x_test1) + noise
x_test1 = x_test1[:, np.newaxis]
```

```python
ridge_model = make_pipeline(PolynomialFeatures(degree = 9), Ridge(alpha=np.exp(-18)))
ridge_model.fit(x_train1,y_train1)

plt.plot(x_train1,ridge_model.predict(x_train1),"red")
plt.plot(x_train1,y_train1,"blue")
plt.show()
```

```python
error=[]
tsterror=[]
ridge_model0 = make_pipeline(PolynomialFeatures(degree = 9), Ridge(alpha=np.exp(-30)))
ridge_model0.fit(x_train1,y_train1)
err0 = sqrt(mean_squared_error(ridge_model0.predict(x_test1), y_test1))
tsterr0 = sqrt(mean_squared_error(ridge_model0.predict(x_train1), y_train1))
error.append(err0)
tsterror.append(tsterr0)

ridge_model1 = make_pipeline(PolynomialFeatures(degree = 9), Ridge(alpha=np.exp(-25)))
ridge_model1.fit(x_train1,y_train1)
err1 = sqrt(mean_squared_error(ridge_model1.predict(x_test1), y_test1))
tsterr1 = sqrt(mean_squared_error(ridge_model1.predict(x_train1), y_train1))
error.append(err1)
tsterror.append(tsterr1)

ridge_model2 = make_pipeline(PolynomialFeatures(degree = 9), Ridge(alpha=np.exp(-20)))
ridge_model2.fit(x_train1,y_train1)
err2 = sqrt(mean_squared_error(ridge_model2.predict(x_test1), y_test1))
tsterr2 = sqrt(mean_squared_error(ridge_model2.predict(x_train1), y_train1))
error.append(err2)
tsterror.append(tsterr2)

ridge_model3 = make_pipeline(PolynomialFeatures(degree = 9), Ridge(alpha=np.exp(-15)))
ridge_model3.fit(x_train1,y_train1)
err3 = sqrt(mean_squared_error(ridge_model3.predict(x_test1), y_test1))
tsterr3 = sqrt(mean_squared_error(ridge_model3.predict(x_train1), y_train1))
error.append(err3)
tsterror.append(tsterr3)

ridge_model4= make_pipeline(PolynomialFeatures(degree = 9), Ridge(alpha=np.exp(-10)))
ridge_model4.fit(x_train1,y_train1)
err4 = sqrt(mean_squared_error(ridge_model4.predict(x_test1), y_test1))
tsterr4 = sqrt(mean_squared_error(ridge_model4.predict(x_train1), y_train1))
error.append(err4)
tsterror.append(tsterr4)

ridge_model5 = make_pipeline(PolynomialFeatures(degree = 9), Ridge(alpha=np.exp(-5)))
ridge_model5.fit(x_train1,y_train1)
err5 = sqrt(mean_squared_error(ridge_model5.predict(x_test1), y_test1))
tsterr5 = sqrt(mean_squared_error(ridge_model5.predict(x_train1), y_train1))
error.append(err5)
tsterror.append(tsterr5)

ridge_model6 = make_pipeline(PolynomialFeatures(degree = 9), Ridge(alpha=np.exp(0)))
ridge_model6.fit(x_train1,y_train1)
err6 = sqrt(mean_squared_error(ridge_model6.predict(x_test1), y_test1))
tsterr6 = sqrt(mean_squared_error(ridge_model6.predict(x_train1), y_train1))
error.append(err6)
tsterror.append(tsterr6)
                            
alpha=[-30,-25,-20,-15,-10,-5,0]
plt.plot(alpha,tsterror,'red')
plt.plot(alpha,error,'blue')
plt.show()
```
